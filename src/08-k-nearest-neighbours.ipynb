```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
```

#### Analisis del dataset

```python
df = pd.read_csv("/workspaces/Emiliano0041-IntroML/data/raw/winequality-red.csv", sep=";")
df
```

- Lo primero que haremos sera convertir los valores de la variable "quality" en los valores que nos pide el proyecto, y ponerlos en una nueva variable llamada label

```python
df["quality"].unique()
```

- Cree una funcion para que le otorgue los valores de 0, 1 y 2 dependiendo el valor de la variable quality para determinar la calidad del vino.

```python
def conversion_q (q):
    if q <= 4:
        return 0
    if q <= 6:
        return 1
    else:
        return 2
    

df["label"] = df["quality"].apply(conversion_q)
```

- Eliminamos la variable quality, ya que ya no me interesa.

```python
df.drop("quality", axis=1, inplace=True)
```

- Usamos value_counts para ver como quedo la distribucion.

```python
df["label"].value_counts()
```

#### Split

- Realizamos el split.

```python
X = df.drop("label", axis=1)
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=22, 
                                                    stratify=y)
```

```python
X_train.head()
```

#### Escalado

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train_norm = scaler.transform(X_train)
X_train_norm = pd.DataFrame(X_train_norm, index = X_train.index, columns=X_train.columns)

X_test_norm = scaler.transform(X_test)
X_test_norm = pd.DataFrame(X_test_norm, index=X_test.index, columns=X_test.columns)
```

```python
modelo_kn = KNeighborsClassifier(n_neighbors=20)
modelo_kn.fit(X_train_norm, y_train)
```

```python
y_pred_1 = modelo_kn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_1)
accuracy
```

```python
report = classification_report(y_test, y_pred_1)
print(report)
```

```python
modelo_kn_2 = KNeighborsClassifier(n_neighbors=1200)
modelo_kn_2.fit(X_train_norm, y_train)
```

```python
y_pred_2 = modelo_kn_2.predict(X_test)
y_pred_2
```

```python
accuracy = accuracy_score(y_test, y_pred_2)
accuracy
```

> Haciendo un analisis y probando distintos hiperparametros, especialmente el de n_neighbors, me di cuenta que el modelo predice todos como clase 1, ya que hay una gran desvalance en el data set en cuanto a los datos de la calidad de los vinos. Lo que yo opino es que en este caso, el modelo de K Nearest Neighbors no es el mas eficiente.
